```python
import numpy as np
import pandas as pd

# Original XOR data
X_original = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_original = np.array([0, 1, 1, 0])

# Generate more samples by adding noise
np.random.seed(42)
n_samples = 1000

# Create noisy versions of each original point
X = []
y = []

for i in range(n_samples):
    # Choose a random original point
    idx = np.random.randint(0, 4)
    original_x = X_original[idx]
    original_y = y_original[idx]
    
    # Add small amount of noise
    noise = np.random.normal(0, 0.1, 2)  # Small noise
    new_x = original_x + noise
    
    # Keep the same label
    X.append(new_x)
    y.append(original_y)

X = np.array(X)
y = np.array(y)

print(f"Expanded dataset size: {X.shape}")
print(f"Class distribution: {np.bincount(y)}")
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Initialize the MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(4,), activation='relu', solver='adam', max_iter=1000, random_state=42)

# Train the model
mlp.fit(X_train, y_train)

# Predict on the test set
y_pred = mlp.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
